```python
import pandas as pd
import re
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud

from nltk import WordNetLemmatizer
lemztr = WordNetLemmatizer()

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

from nltk.tokenize import word_tokenize 
```

```python
reviewlog = pd.read_csv('data/winemag-data-130k-v2.csv')
```

```python
reviewlog
```

```python
reviewlog.description = reviewlog.description.str.lower()
reviewlog.description = reviewlog.description.apply(lambda x: re.sub('\s+', ' ', x))
reviewlog.description = reviewlog.description.apply(lambda x: re.sub('[^\w\s]', '', x))
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
reviewlog.description = reviewlog.description.apply(lambda text: " ".join(word for word in text.split() if word not in stop))
```

```python
for samp in reviewlog.description:
    samp = lemztr.lemmatize(samp)
    samp = stemmer.stem(samp)
```

```python
list_of_varieties = list(reviewlog.variety.value_counts().head().index)
```

```python
for x in list_of_varieties:
    wc = WordCloud(background_color="white", max_words=2000, width=800, height=400)
    wc.generate(' '.join(reviewlog.description[reviewlog.variety == x ]))
    
    print(x)
    plt.figure(figsize=(12, 6))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```
